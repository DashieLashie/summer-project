```python
import pandas as pd
```

```python
tweets = pd.read_parquet("stock_tweets_withsentiment_withemotion_nomerge")
tweets
```

```python


def unify_emotion_features(df: pd.DataFrame) -> pd.DataFrame:

    processed_df = df.copy()


    positive_emotions = ['emotion_joy']
    negative_emotions = ['emotion_anger', 'emotion_disgust', 'emotion_sadness']
    uncertainty_emotions = ['emotion_fear', 'emotion_surprize'] # Note: 'surprize' matches the typo in your notebook


    processed_df['positive_emotion'] = processed_df[positive_emotions].sum(axis=1)
    processed_df['negative_emotion'] = processed_df[negative_emotions].sum(axis=1)
    processed_df['uncertainty_emotion'] = processed_df[uncertainty_emotions].sum(axis=1)


    columns_to_drop = [
        'emotion_anger', 'emotion_disgust', 'emotion_fear',
        'emotion_joy', 'emotion_neutral', 'emotion_sadness', 'emotion_surprize'
    ]


    processed_df = processed_df.drop(columns=columns_to_drop)

    return processed_df

```

```python

tweets_processed = unify_emotion_features(tweets)


print("DataFrame with Unified Emotion Features:")
print(tweets_processed.head())


print("\nNew DataFrame Columns:")
print(tweets_processed.columns)
```
